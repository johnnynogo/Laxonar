```python
import numpy as np
import matplotlib.pyplot as plt

# Constants
N = 8 # Amount of transmitters
pi = np.pi # 3.1415
theta_0 = 0 # Beam angle
v = 343 # Speed of sound [m/s]
f = 40000 # Frequency [Hz]
lambd = v / f
# d = lambd / 2 # Distance between transmitters (measured from center) [m]
d = 0.01 # Distance between transmitters (measured from center) [m]
theta = np.linspace(-90, 90, 1000) * (pi / 180) # Angle in radians

def AF(theta):
    return np.sin( ((N * pi * d) / lambd) * (np.sin(theta) - np.sin(theta_0)) ) / (N * np.sin( (pi * d / lambd) * (np.sin(theta) - np.sin(theta_0)) )) # Array factor

plt.plot(theta, AF(theta))
plt.title("Array Factor")
plt.xlabel("Angle [degrees]")
plt.ylabel("Amplitude [db]")
plt.grid()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Constants
N = 8  # Amount of transmitters
pi = np.pi
theta_0 = 0  # Beam angle in radians
v = 343  # Speed of sound [m/s]
f = 40000  # Frequency [Hz]
lambd = v / f  # Wavelength [m]
# d = lambd / 2  # Distance between transmitters [m]
d = 0.01  # Distance between transmitters (measured from center) [m]

# Angle range in degrees, then convert to radians for calculations
theta_deg = np.linspace(-90, 90, 1000)
theta = theta_deg * (pi / 180)

def array_factor(theta):
    # Calculate arguments for numerator and denominator
    arg_num = (N * pi * d / lambd) * (np.sin(theta) - np.sin(theta_0))
    arg_denom = (pi * d / lambd) * (np.sin(theta) - np.sin(theta_0))
    
    # Handle the case where denominator approaches zero (when theta ≈ theta_0)
    # Using small value epsilon to avoid division by zero
    epsilon = 1e-10
    result = np.zeros_like(theta)
    
    # For angles where denominator is not close to zero
    normal_indices = np.abs(arg_denom) > epsilon
    result[normal_indices] = np.sin(arg_num[normal_indices]) / (N * np.sin(arg_denom[normal_indices]))
    
    # For angles where denominator is close to zero (θ ≈ θ₀)
    # Using the limit value: lim(x→0) sin(Nx)/Nsin(x) = 1
    result[~normal_indices] = 1.0
    
    return result

# Calculate the array factor
AF_values = array_factor(theta)

# Calculate magnitude in dB (20*log10(|AF|))
AF_db = 20 * np.log10(np.abs(AF_values) + 1e-10)  # Adding small constant to avoid log(0)

# Plotting
plt.figure(figsize=(10, 6))

# Plot in dB
plt.plot(theta_deg, AF_db)
plt.title(f"Array Factor (N={N}, d={d:.4f}m, f={f}Hz)")
plt.xlabel("Angle [degrees]")
plt.ylabel("Amplitude [dB]")
plt.grid(True)
plt.xlim(-90, 90)
plt.ylim(-40, 0)  # Set a reasonable dB range

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Constants
pi = np.pi
v = 343  # Speed of sound [m/s]
f = 40000  # Frequency [Hz]
lambd = v / f  # Wavelength [m]
# d = lambd / 2  # Distance between transmitters [m]
d = 0.01

# Define positions for 8 transmitters in honeycomb pattern
# Using hexagonal grid with 4 on top row and 4 on bottom row
# The hexagon side length is d
x_positions = [0, d, 2*d, 3*d,     # Top row
               d/2, 3*d/2, 5*d/2, 7*d/2]  # Bottom row
y_positions = [0, 0, 0, 0,         # Top row
               -d*np.sqrt(3)/2, -d*np.sqrt(3)/2, -d*np.sqrt(3)/2, -d*np.sqrt(3)/2]  # Bottom row

# Convert to numpy arrays
x_pos = np.array(x_positions)
y_pos = np.array(y_positions)

# Shifting to center the array around origin
x_pos = x_pos - np.mean(x_pos)
y_pos = y_pos - np.mean(y_pos)

# Create a meshgrid for evaluation points (viewing angles)
theta = np.linspace(-90, 90, 181) * (pi/180)  # Azimuth angle in radians
phi = np.linspace(-90, 90, 181) * (pi/180)    # Elevation angle in radians
THETA, PHI = np.meshgrid(theta, phi)

# Calculate array factor for the 2D array
def array_factor_2d(theta, phi):
    # Direction cosines
    u = np.sin(theta) * np.cos(phi)
    v = np.sin(phi)
    
    # Initialize array factor
    AF = np.zeros_like(theta, dtype=complex)
    
    # Sum contributions from each transmitter
    for i in range(len(x_pos)):
        phase = 2*pi/lambd * (x_pos[i]*u + y_pos[i]*v)
        AF += np.exp(1j * phase)
    
    # Normalize
    AF = AF / len(x_pos)
    
    return np.abs(AF)

# Calculate array factor for all angles
AF_values = array_factor_2d(THETA, PHI)

# Convert to dB with appropriate floor
AF_dB = 20 * np.log10(AF_values + 1e-10)
AF_dB = np.maximum(AF_dB, -40)  # Limit minimum to -40 dB

# Plot the array layout
plt.figure(figsize=(12, 10))

plt.subplot(221)
plt.scatter(x_pos, y_pos, s=100, c='blue', marker='o')
plt.title('Transmitter Positions (Honeycomb Pattern)')
plt.xlabel('X position [m]')
plt.ylabel('Y position [m]')
plt.grid(True)
plt.axis('equal')

# Connect adjacent transmitters to show honeycomb pattern
for i in range(4):
    plt.plot([x_pos[i], x_pos[i+4]], [y_pos[i], y_pos[i+4]], 'k-', linewidth=1)
    if i < 3:
        plt.plot([x_pos[i], x_pos[i+1]], [y_pos[i], y_pos[i+1]], 'k-', linewidth=1)
        plt.plot([x_pos[i+4], x_pos[i+5]], [y_pos[i+4], y_pos[i+5]], 'k-', linewidth=1)

# Plot 2D array factor in polar form
plt.subplot(222, polar=True)
theta_plot = np.linspace(0, 2*pi, 181)
r_plot = np.abs(array_factor_2d(theta_plot, np.zeros_like(theta_plot)))
plt.plot(theta_plot, r_plot)
plt.title('2D Array Factor (Azimuth Plane)')

# Plot the array factor as 2D color map
plt.subplot(223)
im = plt.pcolormesh(theta*180/pi, phi*180/pi, AF_dB, cmap='jet', shading='auto', vmin=-40, vmax=0)
plt.colorbar(im, label='Amplitude [dB]')
plt.title('Array Factor (dB)')
plt.xlabel('Azimuth Angle [degrees]')
plt.ylabel('Elevation Angle [degrees]')
plt.xlim(-90, 90)
plt.ylim(-90, 90)
plt.grid(True)

# Plot horizontal and vertical cuts
plt.subplot(224)
plt.plot(theta*180/pi, AF_dB[len(phi)//2, :], 'b-', label='Horizontal Cut (φ=0°)')
plt.plot(phi*180/pi, AF_dB[:, len(theta)//2], 'r--', label='Vertical Cut (θ=0°)')
plt.title('Array Factor Cuts')
plt.xlabel('Angle [degrees]')
plt.ylabel('Amplitude [dB]')
plt.grid(True)
plt.xlim(-90, 90)
plt.ylim(-40, 0)
plt.legend()

plt.tight_layout()
plt.show()

# 3D Visualization of the beam pattern
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# Convert to spherical for 3D visualization
r = AF_values
X = r * np.sin(THETA) * np.cos(PHI)
Y = r * np.sin(THETA) * np.sin(PHI)
Z = r * np.cos(THETA)

# Plot 3D surface
surf = ax.plot_surface(X, Y, Z, cmap='jet', alpha=0.8, 
                      linewidth=0, antialiased=False)

ax.set_title('3D Array Factor')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
fig.colorbar(surf, ax=ax, shrink=0.5, aspect=5)

plt.tight_layout()
plt.show()
```
